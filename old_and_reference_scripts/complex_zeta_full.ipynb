```python
# visualise riemann zeta complex function, including for z<1
# tariq rashid, gplv2
```

```python
import mpmath
import numpy
import colorsys
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 2-d array of points
real_points = 400
imag_points = 400
points = numpy.zeros((real_points, imag_points,3), dtype='float64')

# viewport on complex plane
rmin = -15
rmax = 15
imin = -30
imax = 30
```

```python
# calculate zeta(s)

for rp in range(real_points):
    for ip in range(imag_points):
        try:
            real = rmin + ((rmax - rmin) * rp/real_points)
            imag = imin + ((imax - imin) * ip/imag_points)
            z = mpmath.zeta(complex(real,imag))
            z2 = numpy.abs(z)
            rgb = colorsys.hsv_to_rgb(z2, 0.8, 0.8)
            points[ip,rp][0:3] = rgb
        except ValueError:
            # ignore pole
            pass
        pass
    pass
```

```python
# normalise array to [0..1]
points += - points.min()
points /= points.max()
print(points.min(), points.max())
```

```python
# decide how we want to illustrate the complex numbers
#flattened_points = numpy.angle(points)
```

```python
# plot graph, you can right-click and save as an image
plt.figure(figsize = (10,10))
plt.imshow(points, aspect='auto', interpolation='bilinear', origin='lower', extent=[rmin, rmax, imin, imax])
```
